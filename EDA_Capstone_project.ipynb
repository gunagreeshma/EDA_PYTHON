 The dropna() method is used to remove rows or columns with missing (NaN) values from a Pandas

 axis: Specifies whether to remove rows (axis=0) or columns (axis=1) with missing values.

how: Specifies how to determine which rows or columns to drop. It can take the following values:

'any' (default): Drops rows or columns if they contain any missing values.

'all': Drops rows or columns only if all values are missing.

thresh: An integer that specifies the minimum number of non-missing values required for a row or column to be kept. Rows or columns with fewer non-missing values than thresh will be dropped.

subset: A list of column names. If provided, the method will only consider the specified columns when checking for missing values.

inplace: If set to True, it modifies the original DataFrame in place. If set to False (default), it returns a new DataFrame with rows or columns dropped.
